```python
# Importing useful packages
import numpy as np
from scipy import stats
import pandas as pd
import sklearn as sk
import seaborn as sb
import datetime as dt
import pylab 
import matplotlib.pyplot as plt
import pandas.plotting._core as _gfx
%matplotlib inline
```

```python
#This reads in the excel file
ipo_data = pd.read_excel("Competation #1 Raw Data_New.xlsx",header=0,na_values='-' )
#print(ipo_data.dtypes)
```

```python
ipo_data.C2 = ipo_data.C2.astype(bool)
#Changes missing data into the mean
ipo_data=ipo_data.fillna(ipo_data.mean())
# printing summary stats
ipo_data.describe()
```

```python
#df_zscore = (ipo_data - ipo_data.mean())/ipo_data.std()

df = pd.DataFrame({'C1':ipo_data['C1']
                           
                    
                      ,'C3':ipo_data['C3']
                      ,'C4':ipo_data['C4']
                      
                      ,'C5':ipo_data['C5']
                      ,'C6':ipo_data['C6']
                      ,'C7':ipo_data['C7']  
                      ,'T1':ipo_data['T1']
                      ,'T2':ipo_data['T2']
                      ,'T3':ipo_data['T3']
                      ,'T4':ipo_data['T4']
                      ,'S1':ipo_data['S1']
                      ,'S2':ipo_data['S2']
                      ,'S3':ipo_data['S3']})
```

```python
#Calculate Zscore
df['zscore'] = (df['C3'] - df['C3'].mean())/df['C3'].std(ddof=0)
df.describe()
```
